```python
#import these modules
#from nltk.stem import PorterStemmer
#from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from wordcloud import WordCloud, STOPWORDS
stopwords = set(STOPWORDS)


def show_wordcloud(data, title = "Most Popular Discussion About ReactJS"):
    wordcloud = WordCloud(
        background_color='white',
        stopwords=stopwords,
        max_words=30,
        max_font_size=100, 
        scale=3,
        random_state=1 
    ).generate(str(data))

    fig = plt.figure(1, figsize=(12, 12))
    plt.axis('off')
    if title: 
        fig.suptitle(title, fontsize=20)
        fig.subplots_adjust(top=2.3)

    plt.imshow(wordcloud)
    plt.show()

show_wordcloud(data['Body'])
#length = len(stopwords)
#print(f'StopWords length of the dataframe is: {length}')

# ps = PorterStemmer()

# choose some words to be stemmed
# words = ["following","starting","anyone","porting","starting","created length","View","sure","min","render","question","ul","display",
#"page","preloader","Body","Backbone","dtype","help whats","component","responsible","object","app","li","numbers","min","render","Name"]

#for w in words:
	#print(w, " : ", ps.stem(w))
```
